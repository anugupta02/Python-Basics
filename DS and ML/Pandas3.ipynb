```python
import pandas as pd
```

```python
data=pd.read_csv('Book4.csv')
```

```python
x= data.head()
x
```

```python
x=data.iloc[:,0:1]
x.iloc[0]
```

```python
for i in range (len(x)):
    if x.iloc[i,0]%2==0:
        data.iloc[i,2]='Even'
    else:
        data.iloc[i,2]='odd'
```

```python
data.head()
```

```python
l1=[]
l2=[]
for x in range(len(data)):
    if data.iloc[x,2]=='odd':
        l1.append(data.iloc[x,0:3])
    else:
        l2.append(data.iloc[x,0:3])
```

```python
l1=pd.DataFrame(l1)
l2=pd.DataFrame(l2)
```

```python
l1.to_csv('odd.csv')
l2.to_csv('even.csv')
```
